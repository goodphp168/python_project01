```python
score = int(input("請輸入分數："))

if score > 100 or score < 0:
    print("分數輸入錯誤")
elif score >= 90:
    print("成績：A")
elif score >= 80:
    print("成績：B")
elif score >= 70:
    print("成績：C")
else:
    print("成績：D")
```
